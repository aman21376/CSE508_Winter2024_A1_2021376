```python
import os
import shutil

fpath = "text_files"
dpath = "processed_files"
if not os.path.exists(dpath):
    os.makedirs(dpath)

for fname in os.listdir(fpath):
    source_file_path = os.path.join(fpath, fname)

    if os.path.isfile(source_file_path):
        with open(source_file_path, 'r') as source_file:
            content = source_file.read()
        content_lower = content.lower()
        
        # Full path to the destination file
        destination_file_path = os.path.join(dpath, fname)
        
        with open(destination_file_path, 'w') as destination_file:
            destination_file.write(content_lower)

        print(f"Processed file: {source_file_path} | Saved to: {destination_file_path}")
```
